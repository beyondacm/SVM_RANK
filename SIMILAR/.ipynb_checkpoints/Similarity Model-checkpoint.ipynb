```python
import graphlab
import numpy as np
import pandas as pd
import matplotlib 
```

```python
best_answer = graphlab.SFrame.read_csv('best_answer.csv')
```

```python
best_answer
```

```python
best_answer['text'] = best_answer['answer_content'] + best_answer['question_tags'] + best_answer['question_content']
```

```python
# best_answer[0]
best_answer['word_count'] = graphlab.text_analytics.count_words(best_answer['text'])
```

```python
tfidf = graphlab.text_analytics.tf_idf(best_answer['word_count'])
```

```python
tfidf
```

```python
best_answer['tfidf'] = tfidf
```

```python
best_answer[0]
```

```python
knn_model = graphlab.nearest_neighbors.create(best_answer, features=['tfidf'], label='question_id')
```
